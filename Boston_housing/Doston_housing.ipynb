# Linear Regression with Scikit-Learn

```python
df=pd.read_csv("Boston_housing.data", delim_whitespace=True,header=None)
```

```python
col_name=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV'] 
```

```python
df.columns=col_name 
```

```python
df.head()
```

```python
# 1. Input values for x and y
x=df['RM'].values.reshape(-1,1)
```

```python
y=df['MEDV'].values
```

```python
# 2. import the model
from sklearn.linear_model import LinearRegression
```

```python
# 3. Instantiate model
model= LinearRegression()
```

```python
# 4. Train the model
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,8));
sns.regplot(x=x,y=y);
plt.xlabel('Average number of room per dwelling')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.show()
```

```python
import numpy as np 
model.predict(np.array([8]).reshape(1,-1))      
```

# Assumption of Regression Analysis
##### 1. Linear Relationship between the features and target: 
##### 2. Little or no Multicollinearity between the features :Multicollinearity is a state of very high inter-correlations or inter-associations among the independent variables. It is therefore the type of disturbance in the data present weakens the statistiacal power of the regression model. Pair plots and heatmaps(correlation matrix) can be used for identifying highly correlated features.
##### 3. Low or no Hetrosecdasticity: Errors terms should not explain other error term behaver when ploted scattered plot they should show ramdom sprade instaded showing any pattern
##### 4. Normal Distribuation of the residuals can be validated by plotting a q-q plot:
##### 5. Little or No autocorrelation in the residuals:

```python
# 3. !pip install yellowbrick
from yellowbrick.regressor import ResidualsPlot
plot_visuals=ResidualsPlot(LR1)

plot_visuals.fit(X_train,y_train)
plot_visuals.score(X_test,y_test)

plt.legend(loc='upper left')
plot_visuals.show()
```

```python
# 4. Normal Distribuation of the residuals can be validated by plotting a q-q plot 
plot_visuals=ResidualsPlot(LR1,hist=False,qqplot=True) 
plot_visuals.fit(X_train,y_train) 
plot_visuals.score(X_test,y_test) 
plot_visuals.show() 
```
